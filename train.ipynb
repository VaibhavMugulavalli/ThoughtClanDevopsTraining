```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix

df=pd.read_csv(r"C:\Users\vaibh\OneDrive\Documents\ThoughtClan\CI\CD\credit_customers.csv")
df
```

```python
features=['credit_history','credit_amount','age','existing_credits']
target='class'

df=df[features+[target]]
df.dropna(inplace=True)
df
```

```python
import joblib
le=LabelEncoder()
df['credit_history']=le.fit_transform(df['credit_history'])
joblib.dump(le,'credit_history_encoded.pkl')
df
```

```python
X=df[features]
Y=df[target]

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)

model=DecisionTreeClassifier(max_depth=5,random_state=42)
model.fit(X_train,Y_train)

joblib.dump(model, 'model.pkl')
```
